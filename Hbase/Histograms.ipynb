```python
f = open('/home/vagrant/data_c.txt')
```

```python
data =[]
for row in f:
    data.append(row.strip())
```

```python

dis = list(set(data))
dis = sorted(dis)
dis_i=[]
for i in dis:
    dis_i.append(int(i))
dis_i
```

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import *
import matplotlib.axes as axes
```

```python
data_a = np.asarray(data,dtype=int)
```

```python
data_a
```

```python
plt.hist(data_a, bins=np.arange(data_a.min(), data_a.max()+1),facecolor='red', edgecolor='gray')
plt.title("Histogram for crazy credit")
plt.xlabel("crazy credit count per movie")
plt.ylabel("frequency of count of credit per movie")
plt.xticks(dis_i,rotation=70)
```
